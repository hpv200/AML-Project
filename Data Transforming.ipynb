```python
import pandas as pd
import datetime
from datetime import timedelta
import numpy as np
```

```python
flights_test = pd.read_csv('test.csv')
flights = pd.read_csv('train.csv')
```

## Clean entries not present in test set 

```python
print('ORIGIN_AIRPORT_train: ', len(flights.ORIGIN_AIRPORT.unique()))
print('AIRLINE_train: ', len(flights.AIRLINE.unique()))
print('ORIGIN_AIRPORT_test: ', len(flights_test.ORIGIN_AIRPORT.unique()))
print('AIRLINE_test: ', len(flights_test.AIRLINE.unique()))
```

```python
orig_airport_test = set(flights_test.ORIGIN_AIRPORT.unique())
airline_test = set(flights_test.AIRLINE.unique())
orig_airport = set(flights.ORIGIN_AIRPORT.unique())
airline = set(flights.AIRLINE.unique())
diff_airport = list(orig_airport ^ orig_airport_test)
diff_airline = list(airline_test ^ airline)
```

```python
flights = flights[flights.AIRLINE != 'US'].reset_index(drop=True)
```

```python
for x in range(len(diff_airport)):
    flights = flights[flights.ORIGIN_AIRPORT != diff_airport[x]]

flights = flights.reset_index(drop=True)
```

## One-hot encoding

```python
one_hot = pd.get_dummies(flights.DAY_OF_WEEK, prefix = 'day_of_week')
flights = flights.drop('DAY_OF_WEEK',axis = 1)
flights = flights.join(one_hot)

one_hot = pd.get_dummies(flights.ORIGIN_AIRPORT, prefix = 'ORIGIN_AIRPORT')
flights = flights.drop('ORIGIN_AIRPORT',axis = 1)
flights = flights.join(one_hot)

one_hot = pd.get_dummies(flights.AIRLINE, prefix = 'AIRLINE')
flights = flights.drop('AIRLINE',axis = 1)
flights = flights.join(one_hot)

one_hot = pd.get_dummies(flights.TIME_OF_DAY, prefix = 'TIME_OF_DAY')
flights = flights.drop('TIME_OF_DAY',axis = 1)
flights = flights.join(one_hot)
```

```python
flights = flights.drop(columns = ['SCHEDULED_DEPARTURE_TIME', 'Month'])
```
